```
#File location and type
file_location = "dbfs:/mnt/s3data/HIGGS.csv"
file_type ="csv"

#CSV Options
infer_schema = "true"
first_row_is_header = "false"
delimiter = ","

#The applied options are for CSV files.
sdf = spark.read.format(file_type) \
  .option("multiline", "true") \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(sdf)
```

```
#Change column headers
from functools import reduce
oldColumns = sdf.schema.names
newColumns =['label','feature1','feature2','feature3','feature4','feature5','feature6','feature7','feature8','feature9','feature10','feature11','feature12','feature13','feature14','feature15','feature16','feature17','feature18','feature19','feature20','feature21','feature22','feature23','feature24','feature25','feature26','feature27','feature28']

df = reduce(lambda sdf, idx: sdf.withColumnRenamed(oldColumns[idx], newColumns[idx]), range(len(oldColumns)), sdf)
df.printSchema()
df.show()
```

```
#Write to parquet 
df.write.parquet("/tmp/output/higgsdf.parquet")
```

```
#Check if parquet is readable
#Read parquet 

data = spark.read.parquet("/tmp/output/higgsdf.parquet")
data.show()
```
